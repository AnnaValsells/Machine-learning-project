```python
# Función para localizar estos valores vacios:

def is_blank_or_whitespace(cell):
    return isinstance(cell, str) and cell.strip() == ''

mask = data.applymap(is_blank_or_whitespace)
empty_rows = data[mask.any(axis=1)]

empty_rows
```

```python
# Relleno los espacios vacios con la palabra 'unknown'

data = data.applymap(lambda x: 'unknown' if is_blank_or_whitespace(x) else x)
```
